```python
# Initialize a marker cluster
marker_cluster = MarkerCluster()

# Iterate through the spacex_df DataFrame
for index, record in spacex_df.iterrows():
    # Define the coordinate for the marker
    coordinates = [record['Lat'], record['Long']]
    
    # Define the marker's icon color based on the 'marker_color' column
    marker_color = record['marker_color']
    
    # Create a Marker object and add it to the marker cluster
    marker = folium.Marker(
        location=coordinates,
        icon=folium.Icon(color='white', icon_color=marker_color),
        popup=f"Launch Site: {record['Launch Site']}<br>Outcome: {record['class']}",
    )
    marker_cluster.add_child(marker)

# Add the marker cluster to the site_map
site_map.add_child(marker_cluster)

# Display the updated site_map
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# Function to calculate the Haversine distance between two coordinates
def calculate_distance(lat1, lon1, lat2, lon2):
    # Radius of the Earth in km
    R = 6371.0
    
    # Convert latitude and longitude from degrees to radians
    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)
    
    # Differences in coordinates
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    
    # Haversine formula
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    
    # Calculate the distance
    distance = R * c
    
    return distance

# Launch site coordinates (example)
launch_site_lat = 28.5721
launch_site_lon = -80.6480

# Coordinates of the closest coastline (example)
closest_coastline_lat = 28.56367
closest_coastline_lon = -80.57163

# Calculate the distance between the launch site and the closest coastline
distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, closest_coastline_lat, closest_coastline_lon)

print(f"Distance to closest coastline: {distance_coastline} km")
```

```python
from folium import plugins

# Calculate the distance between the launch site and the closest coastline
distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, closest_coastline_lat, closest_coastline_lon)

# Create a folium.Marker to show the distance
distance_marker = folium.Marker(
    [closest_coastline_lat, closest_coastline_lon],
    icon=plugins.BeautifyIcon(
        icon_shape='circle',
        border_color='blue',
        border_width=2,
        number=round(distance_coastline, 2),
        background_color='transparent'
    )
)

# Add the distance_marker to the site_map
site_map.add_child(distance_marker)

# Display the map
site_map


```

```python
# Define the coordinates for the PolyLine
coordinates = [(launch_site_lat, launch_site_lon), (closest_coastline_lat, closest_coastline_lon)]

# Create a folium.PolyLine object
lines = folium.PolyLine(locations=coordinates, weight=1)

# Add the PolyLine to the site_map
site_map.add_child(lines)

# Display the map
site_map
```
