```python
# andas is a software library written for the Python programming language for data manipulation and analysis.
import pandas as pd
#NumPy is a library for the Python programming language, adding support for large, multi-dimensional arrays and matrices, along with a large collection of high-level mathematical functions to operate on these arrays
import numpy as np
# Matplotlib is a plotting library for python and pyplot gives us a MatLab like plotting framework. We will use this in our plotter function to plot data.
import matplotlib.pyplot as plt
#Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics
import seaborn as sns
```

```python
df=pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_2.csv")
```

```python
df.head(5)
```

```python
sns.catplot(y="PayloadMass", x="FlightNumber", hue="Class", data=df, aspect = 5)
plt.xlabel("Flight Number",fontsize=20)
plt.ylabel("Pay load Mass (kg)",fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(12, 8))  

sns.scatterplot(data=df, x="FlightNumber", y="LaunchSite", hue="Class", palette="Set2")

# Set plot labels and title
plt.xlabel("Flight Number")
plt.ylabel("Launch Site")
plt.title("Scatter Plot of Flight Number vs. Launch Site (Colored by Class)")

# Show the legend
plt.legend(title="Class", loc="upper right")

plt.show()
```

```python


plt.figure(figsize=(12, 8))  


sns.scatterplot(data=df, x="PayloadMass", y="LaunchSite", hue="Class", palette="Set2")

# Set plot labels and title
plt.xlabel("Payload Mass (kg)")
plt.ylabel("Launch Site")
plt.title("Scatter Plot of Payload Mass vs. Launch Site (Colored by Class)")

# Show the legend
plt.legend(title="Class", loc="upper right")

# Show the plot
plt.show()
```

```python
# Group the data by the "Orbit" column and calculate the mean of the "Class" column
orbit_class_means = df.groupby("Orbit")["Class"].mean()

# Display the result
print(orbit_class_means)
```

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(12, 6))  # Set the figure size

# Plot the bar chart
plt.bar(orbit_class_means.index, orbit_class_means.values, color='skyblue')

# Set plot labels and title
plt.xlabel("Orbit")
plt.ylabel("Rate of Success")
plt.title("Rate of Success for Each Orbit")

# Rotate x-axis labels for better visibility (optional)
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
 
year = []  # Create an empty list to store years

def Extract_year():
    for date in df["Date"]:
        year.append(date.split("-")[0])
    return year

# Call the function to extract years from the "Date" column
years = Extract_year()

# Print the list of years
print(years)

    
```

```python
features = df[['FlightNumber', 'PayloadMass', 'Orbit', 'LaunchSite', 'Flights', 'GridFins', 'Reused', 'Legs', 'LandingPad', 'Block', 'ReusedCount', 'Serial']]
features.head()
```

```python
# List of columns to one-hot encode
columns_to_encode = ['Orbit', 'LaunchSite', 'LandingPad', 'Serial']

# Apply one-hot encoding using get_dummies
features_one_hot = pd.get_dummies(features, columns=columns_to_encode)

# Display the resulting DataFrame
features_one_hot.head()
```

```python
# Cast all numeric columns to float64
features_one_hot = features_one_hot.astype(float)

# Display the resulting DataFrame
features_one_hot.head()
```

```python
features_one_hot.to_csv('dataset_part_3.csv', index=False)
```
